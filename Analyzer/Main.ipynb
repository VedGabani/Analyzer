```python
import numpy as np

def array():
    while True:
        print("\nArray function is running...\n")
        print("1. Create an array")
        print("2. Display the array")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            print("1. one-dimensional array\n")
            print("2. two-dimensional array\n")
            print("3. three-dimensional array\n")
            array_choice = input("Enter your choice: ")
            if array_choice == "1":
                
                user_input = input("Enter 1-D array elements separated by spaces: ")
                num_list = [int(x) for x in user_input.split()]
                arr_1d = np.array(num_list)
                print("1-D array created successfully!")

            elif array_choice == "2":

                rows = int(input("Enter the number of rows: "))
                cols = int(input("Enter the number of columns: "))
                arr_2d = np.zeros((rows, cols))
                print("2-D array created successfully!")
                
            elif array_choice == "3":

                depth = int(input("Enter the depth: "))
                rows = int(input("Enter the number of rows: "))
                cols = int(input("Enter the number of columns: "))
                arr_3d = np.zeros((depth, rows, cols))
                print("3-D array created successfully!")

            else:

                print("Invalid choice. Please try again.")

        elif choice == "2":
            try:
                print("1-D array:", arr_1d)
            except NameError:
                print("1-D array not created yet.")
            try:
                print("2-D array:\n", arr_2d)
            except NameError:
                print("2-D array not created yet.")
            try:
                print("3-D array:\n", arr_3d)
            except NameError:
                print("3-D array not created yet.")

        elif choice == "3":
            print("Exiting the array function.")
            break

        else:
            print("Invalid choice. Please try again.")
```

```python
def combine():
    print("\nCombine function is running...\n")
    print("1. Combine two arrays")
    print("2. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":

        user_input1 = input("Enter elements of the first array separated by spaces: ")
        num_list1 = [int(x) for x in user_input1.split()]
        arr1 = np.array(num_list1)

        user_input2 = input("Enter elements of the second array separated by spaces: ")
        num_list2 = [int(x) for x in user_input2.split()]
        arr2 = np.array(num_list2)

        combined_array = np.concatenate((arr1, arr2))
        print("Combined array:", combined_array)

    elif choice == "2":
        print("Exiting the combine function.")
        break
    
    else:
        print("Invalid choice. Please try again.")
```
