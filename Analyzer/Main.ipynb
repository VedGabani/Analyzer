```python
import numpy as np
```

```python

global arr_1d, arr_2d, arr_3d

def array_creation():
    while True:
        print("\nArray function is running...\n")
        print("1. Create an array")
        print("2. Display the array")
        print("3. Exit")
        choice = input("Enter your choice -_- ")

        if choice == "1":
            print("1. one-dimensional array\n")
            print("2. two-dimensional array\n")
            print("3. three-dimensional array\n")
            array_choice = input("Enter your choice -_- ")
            if array_choice == "1":
                
                user_input = input("Enter 1-D array elements separated by spaces -_- ")
                num_list = [int(x) for x in user_input.split()]
                arr_1d = np.array(num_list)
                print("1-D array created successfully!")

            elif array_choice == "2":

                rows = int(input("Enter the number of rows -_- "))
                cols = int(input("Enter the number of columns -_- "))
                arr_2d = np.zeros((rows, cols))
                print("2-D array created successfully!")
                
            elif array_choice == "3":
                
                depth = int(input("Enter the depth -_- "))
                rows = int(input("Enter the number of rows -_- "))
                cols = int(input("Enter the number of columns -_- "))
                arr_3d = np.zeros((depth, rows, cols))
                print("3-D array created successfully!")

            else:

                print("Invalid choice. Please try again.")

        elif choice == "2":
            try:
                print("1-D array -_- ", arr_1d)
            except NameError:
                print("1-D array not created yet.")
            try:
                print("2-D array -_-:\n", arr_2d)
            except NameError:
                print("2-D array not created yet.")
            try:
                print("3-D array -_-:\n", arr_3d)
            except NameError:
                print("3-D array not created yet.")

        elif choice == "3":
            print("Exiting the array function.")
            break

        else:
            print("Invalid choice. Please try again.")
```

```python
def combine():
    while True:
        print("\nCombine function is running...\n")
        print("1. Combine two arrays")
        print("2. Exit")

        choice = input("Enter your choice -_- ")

        if choice == "1":

            user_input1 = input("Enter elements of the first array separated by spaces -_- ")
            num_list1 = [int(x) for x in user_input1.split()]
            arr1 = np.array(num_list1)

            user_input2 = input("Enter elements of the second array separated by spaces -_- ")
            num_list2 = [int(x) for x in user_input2.split()]
            arr2 = np.array(num_list2)

            combined_array = np.concatenate((arr1, arr2))
            print("Combined array:", combined_array)

        elif choice == "2":
            print("Exiting the combine function.")
            break

        else:
            print("Invalid choice. Please try again.")
```

```python
def math_operations():
    while True:
        print("\nMath function is running...\n")
        print("1. Perform mathematical operations on an array")
        print("2. Exit")

        choice = input("Enter your choice -_- ")

        if choice == "1":

            user_input = input("Enter array elements separated by spaces -_- ")
            num_list = [int(x) for x in user_input.split()]
            arr = np.array(num_list)

            print("Array:", arr)
            print("1. Sum of elements")
            print("2. Mean of elements")
            print("3. Standard deviation of elements")
            math_choice = input("Enter your choice -_- ")

            if math_choice == "1":
                print("Sum of elements -_-", np.sum(arr))
            elif math_choice == "2":
                print("Mean of elements -_- ", np.mean(arr))
            elif math_choice == "3":
                print("Standard deviation of elements -_-", np.std(arr))
            else:
                print("Invalid choice. Please try again.")

        elif choice == "2":
            print("Exiting the math function.")
            break

        else:
            print("Invalid choice. Please try again.")
```

```python
def stats():
    while True:
        print("\nStats function is running...\n")
        print("1. Calculate statistics of an array")
        print("2. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":

            user_input = input("Enter array elements separated by spaces: ")
            num_list = [int(x) for x in user_input.split()]
            arr = np.array(num_list)

            print("Array:", arr)
            print("1. Minimum value")
            print("2. Maximum value")
            print("3. Median value")
            stats_choice = input("Enter your choice: ")

            if stats_choice == "1":
                print("Minimum value:", np.min(arr))
            elif stats_choice == "2":
                print("Maximum value:", np.max(arr))
            elif stats_choice == "3":
                print("Median value:", np.median(arr))
            else:
                print("Invalid choice. Please try again.")

        elif choice == "2":
            print("Exiting the stats function.")
            break

        else:
            print("Invalid choice. Please try again.")
```

```python
def search():
    while True:
        print("\nSearch function is running...\n")
        print("1. Search for an element in an array")
        print("2. Exit")

        choice = input("Enter your choice -_- ")

        if choice == "1":

            user_input = input("Enter array elements separated by spaces -_- ")
            num_list = [int(x) for x in user_input.split()]
            arr = np.array(num_list)

            search_element = int(input("Enter the element to search for -_- "))
            if search_element in arr:
                print(f"Element {search_element} found in the array.")
            else:
                print(f"Element {search_element} not found in the array.")

        elif choice == "2":
            print("Exiting the search function.")
            break

        else:
            print("Invalid choice. Please try again.")
```

```python
def main():
    while True:
        print("\nMain Menu\n")
        print("1. Array Operations")
        print("2. Combine Arrays")
        print("3. Mathematical Operations")
        print("4. Statistical Operations")
        print("5. Search in Array")
        print("6. Exit")

        main_choice = input("Enter your choice: ")

        if main_choice == "1":array_creation()
        elif main_choice == "2":combine()
        elif main_choice == "3":math_operations()
        elif main_choice == "4":stats()
        elif main_choice == "5":search()
        elif main_choice == "6":
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

print("Welcome to the Array Operations Program!")

main()
```
