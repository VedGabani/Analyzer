```python
import numpy as np

def create_array():
    print("1. 1D\n2. 2D")
    ch = int(input("Choice: "))

    if ch == 1:
        a = list(map(int, input("Enter: ").split()))
        return np.array(a)

    elif ch == 2:
        r = int(input("Rows: "))
        c = int(input("Cols: "))
        a = list(map(int, input("Enter: ").split()))
        return np.array(a).reshape(r, c)
```
