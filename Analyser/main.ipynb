```python
import numpy as np

# ===== ARRAY =====
def create_array():
    print("1. 1D\n2. 2D")
    ch = int(input("Choice: "))

    if ch == 1:
        a = list(map(int, input("Enter: ").split()))
        return np.array(a)

    elif ch == 2:
        r = int(input("Rows: "))
        c = int(input("Cols: "))
        a = list(map(int, input("Enter: ").split()))
        return np.array(a).reshape(r, c)

# ===== MATH =====
def math_op(arr):
    print("1.Add 2.Sub 3.Mul 4.Div")
    ch = int(input("Choice: "))

    b = list(map(int, input("Enter second array: ").split()))
    arr2 = np.array(b).reshape(arr.shape)

    if ch == 1:
        print(arr + arr2)
    elif ch == 2:
        print(arr - arr2)
    elif ch == 3:
        print(arr * arr2)
    elif ch == 4:
        print(arr / arr2)

# ===== COMBINE =====
def combine(arr):
    b = list(map(int, input("Enter second array: ").split()))
    arr2 = np.array(b).reshape(arr.shape)
    print(np.vstack((arr, arr2)))

def split(arr):
    print(np.array_split(arr, 2))

# ===== SEARCH =====
def search_sort(arr):
    print("1.Search 2.Sort")
    ch = int(input("Choice: "))

    if ch == 1:
        x = int(input("Enter value: "))
        print(np.where(arr == x))
    elif ch == 2:
        print(np.sort(arr))

# ===== STATS =====
def stats(arr):
    print("1.Sum 2.Mean 3.Median")
    ch = int(input("Choice: "))

    if ch == 1:
        print(np.sum(arr))
    elif ch == 2:
        print(np.mean(arr))
    elif ch == 3:
        print(np.median(arr))

# ===== MAIN =====
arr = None

while True:
    print("\n1.Create\n2.Math\n3.Combine\n4.Split\n5.Search\n6.Stats\n7.Exit")
    ch = int(input("Choice: "))

    if ch == 1:
        arr = create_array()
        print(arr)

    elif ch == 2 and arr is not None:
        math_op(arr)

    elif ch == 3 and arr is not None:
        combine(arr)

    elif ch == 4 and arr is not None:
        split(arr)

    elif ch == 5 and arr is not None:
        search_sort(arr)

    elif ch == 6 and arr is not None:
        stats(arr)

    elif ch == 7:
        break

    else:
        print("Create array first!")
```
